```python
# Sorting Dictionary
record={
    'z':80,
    'a':45,
    'c':90,
    'b':74,
    'e':65
}
# {a:45,b:74,c:90,e:65,z=80}
print(dict(sorted(record.items())))
```

```python
record=[('a', 45), ('b', 74), ('c', 90), ('e', 65), ('z', 80)]
print(dict(record))
```

```python
# Sorting Dictionary
record={
    'z':80,
    'a':45,
    'c':90,
    'b':74,
    'e':65
}
# {a:45,b:74,c:90,e:65,z=80}
# print(dict(sorted(record.items())))
reversed_items=[(v,k) for k,v in record.items()]
```

```python
print(reversed_items)
```

```python
sorted_revered=sorted(reversed_items)
print(sorted_revered)
```

```python
result={k:v for v,k in sorted_revered}
print(result)
```

```python
result=dict(sorted(record.items(),key=lambda x:x[1]))
print(result)
```
