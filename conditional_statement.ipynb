```python
name=input("Enter Your Name: ")
unit=int(input("Enter Your Electicity Unit: "))
customer_type=input("Enter Customer Type: ")
meter_charge=100
if unit<=100:
    bill=unit*5
elif unit<=300:
    bill=(100*5)+((unit-100)*7)
else:
    bill=(100*5)+(200*7)+((unit-300)*10)

# Enerygy Calculations

if bill>2000:
    disccount5=bill*0.05
    payble_amount=bill-disccount5

print("===============ENERGY BILL CALCULATOR=====================")
print(f"Customer Name is {name}")
print(f"Total Electicity Unit{unit}")
print(f"Customer Type {customer_type}")
print(f"Discount 5% {disccount5}")
print(f"Total Electricity Bill {bill}")
print(f"Final Payble Bill {payble_amount+meter_charge}")
```
