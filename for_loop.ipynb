```python
3.Online Shopping Order Summary
Problem
A customer buys 5 products.
Using a for loop:
Enter the price of each product.
Find:
Total bill
Most expensive product price
Cheapest product price
Number of products costing more than ₹1000
Example
Input
1200
450
3000
999
1500
Output
Total Bill : 7149
Highest Price : 3000
Lowest Price : 450
Products Above ₹1000 : 3
```

```python
# Online Shopping Order Summary
prices = []
total = 0
highest = 0
lowest = 0
count_above_1000 = 0
print("Enter Price of 5 Products\n")
for i in range(5):
    price = float(input(f"Enter Price of Product {i+1}: "))
    prices.append(price)
    total += price
    if i == 0:
        highest = price
        lowest = price
    else:
        if price > highest:
            highest = price
        if price < lowest:
            lowest = price
    if price > 1000:
        count_above_1000 += 1
print("\n========== Shopping Order Summary ==========")
print("\nProduct Prices:")
for i in range(5):
    print(f"Product {i+1} : {prices[i]}")
print("\nTotal Bill : ", total)
print("Highest Price :", highest)
print("Lowest Price : ", lowest)
print("\nProducts Costing More Than 1000:")
for i in range(5):
    if prices[i] > 1000:
        print(f"Product {i+1} : {prices[i]}")
print(f"\nNumber of Products Above 1000 : {count_above_1000}")
```
