# While Loop

A While Loop executes a block of code of as long as the given condition is True:

# Syntax
while conditon:
    # Code

```python
# print( 1 to 5):
for i in range(1,6):
    print(i)
```

```python
i=1
while i<=5:
    print(i)
    i+=1
    
```

```python
num=2
while num<=20:
    print(num)
    num+=2
```

```python
num=20
while num>=2:
    print(num)
    num-=2
```

```python
i=1
while i<=20:
    if i==15:
        break
    print(i)
    i+=1
```

```python
i=0
while i<20:
    i+=1
    if i in [6,7]:
        continue
    print(i)
```

```python
i=1
while i<=5:
    print(i)
    i+=1
else:
    print("Your While Loop is Completed")
```

```python
correct_pin=9661
pin=int(input("Enter Your PIN "))
attempts=0
while pin!=correct_pin and attempts<=3:
    print("Wrong PIN Entereted Please Try Again")
    pin=int(input("Enter Your PIN "))
    attempts+=1
print("Login Succesfully")
```

```python
correct_pin=9661
pin=int(input("Enter Your PIN "))
attempts=0
while pin!=correct_pin and attempts<3:
    print("Wrong PIN Entereted Please Try Again")
    pin=int(input("Enter Your PIN "))
    attempts+=1
if pin==correct_pin:
    print("Login Successfully")
else:
    print("Account Has Been Blocked")
```

```python
correct_pin=9661
attempts=0
max_attempts=4
while attempts<max_attempts:
    pin=int(input("Enter Your PIN "))
    if pin==correct_pin:
        print("Login Succesfully")
        break
    else:
        attempts+=1
        print("Wrong Pin Entered")
if attempts==max_attempts:
    print("Your Account has been blocked")
```

```python
n=int(input("Enter Your NUmber"))
total=0
i=2
while i<=n:
    total+=i
    i+=2
print(f"The Final Sum if Your Ented Number is {total}")
```

```python
n=int(input("Enter Your NUmber"))
fact=1
i=1
while i<=n:
    fact*=i
    i+=1
print(f"The Final Factorial of  Your Ented Number is {fact}")
```
