```python
# Tuple Packing

p=45,100,85,96
print(p)
```

```python
# Unpacking
s1,s2,s3,s4=p
```

```python
print(s1)
print(s2)
print(s3)
print(s4)
```

```python
for i in p:
    print(i)
```
